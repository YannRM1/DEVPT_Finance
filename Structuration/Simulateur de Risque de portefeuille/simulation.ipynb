```python
def prix_obligation(coupons, r, nominal, T):
    prix = 0
    for t in range(1,T+1):
        prix += coupons[t-1]/(1+r)**t
    prix += nominal/(1+r)**T
    return prix 

coupons = [5,5,5,5,105]
r = 0.03
nominal = 100
T = 5
prix = prix_obligation(coupons, r, nominal, T)
print(f"Le prix de l'obligation est {prix}")
```

```python
def duration(coupons, r, T):
    duree = 0
    for t in range(1,T+1):
        duree += (t*coupons/(1+r)**t)/prix
    return duree 
```
